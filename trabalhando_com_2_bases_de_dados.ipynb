```python
import pandas as pd
import seaborn as sns
```

```python
vendas= pd.read_excel("/content/drive/MyDrive/DNC/Formação de Dados/colab/trabalhando com 2 bases de dados/varejo.xlsx")
vendas.head ()
```

```python
vendas.info ()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"]= vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["Nome_Departamento"]= vendas["Nome_Departamento"].str.replace(" ", "_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas["estado"].fillna("MS", inplace=True)
vendas.info ()
```

```python
preco_medio= vendas["Preço"].mean ()
vendas["Preço"].fillna(preco_medio, inplace=True)
vendas.info ()
```

```python
vendas[["Preço", "Preço_com_frete"]].describe ()
```

```python
vendas_preço_ok= vendas.query ("Preço < Preço_com_frete")
vendas_preço_ok.describe ()
```

```python
vendas_preço_ok.groupby ("Nome_Departamento").idcompra.nunique().sort_values (ascending=False).reset_index ().head ()
```

```python
departXpreco= round (vendas_preço_ok.groupby ("Nome_Departamento")["Preço_com_frete"].agg("mean"), 2).sort_values (ascending=False).reset_index ()
departXpreco.head (10)
```

```python
cliente= pd.read_excel ("/content/drive/MyDrive/DNC/Formação de Dados/colab/trabalhando com 2 bases de dados/cliente_varejo.xlsx")
cliente.info ()
```

```python
cliente.describe ()
```

```python
cliente= cliente.astype({"renda":"float"})
cliente.info ()
```

```python
vendas_cliente= vendas_preço_ok.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head ()
```

```python
canalXrenda= round (vendas_cliente.groupby ("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False), 2).reset_index ()
sns.barplot(data=canalXrenda, x="idcanalvenda", y="renda")
```

```python
bandeiraXidade= round (vendas_cliente.groupby ("bandeira")["idade"].agg("mean").sort_values(ascending=False), 2).reset_index ()
sns.barplot(data=bandeiraXidade, x="bandeira", y="idade")
```

```python
import matplotlib.pyplot as plt
fig= plt.figure (figsize=(10,8))
plt.bar (bandeiraXidade["bandeira"], bandeiraXidade["idade"], color="green")
plt.ylabel ("Média de idade")
plt.title ("Idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color="black")
```
